# EXPLORATORY DATA ANALYSIS

**Extract The Data When No. of Bedrooms And No. of Bathrooms is Greater than or equal 5:**

```python
ef=df.loc[((df['No. of Bedrooms']>=5) | (df['No. of Bathrooms']>=5)) & (df['Price']<2000000)]
ef
```

**Show The Bar Graph**

```python
plt.figure(figsize=(25,8))
plt.bar(ef['Property Name'],ef['Price'])
plt.xlabel('Property Name',size=20)
plt.ylabel('Price',size=20)
plt.title('Greater Than or equal to 5BHK Property Price',size=20)
plt.xticks(ef['Property Name'],rotation=90,size=8)
plt.show()
```

**Flat/Apartment or Studio Data By Affordable Price**

**Get The Minimum Price By House Type**

```python
sdf=df.groupby('House Type')['Price'].min().sort_values()
sdf
```

```python
sdf=df.loc[((df['House Type']=='Flat / Apartment') | (df['House Type']=='Studio') | (df['House Type']=='New development')) & (df['Price']<370000)]
sdf['Price'].sort_values(ascending=True)
```

**Plot The Data With Location**

```python
plt.figure(figsize=(20,8))
plt.scatter(sdf['Property Name'],sdf['Location'])
plt.xlabel('Property Name',size=30)
plt.ylabel('Location',size=30)
plt.xticks(sdf['Property Name'],rotation=90,size=15)
plt.title('Flat/Apartment Or Studio Location Graph',size=40)
plt.show()
```

**Find The Property Data By maximum area**

**Get The Maximum Area By House Type**

```python
edf=df.groupby('House Type')['Area in sq ft'].max().sort_values()
edf
```

```python
edf=df.loc[((df['House Type']=='Flat / Apartment') | (df['House Type']=='New development')) & (df['Area in sq ft']>5000)]
edf
```

**Show The plot Graph**

```python
plt.plot(edf['Property Name'],edf['Area in sq ft'])
plt.xlabel('Property Name',size=15)
plt.ylabel('Area in sq ft',size=15)
plt.xticks(edf['Property Name'],rotation=60)
plt.title('Flat/Apartment or New development Data By maximum area')
plt.show
```
